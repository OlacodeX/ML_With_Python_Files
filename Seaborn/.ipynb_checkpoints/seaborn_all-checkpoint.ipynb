### Dist plot

### Joint plot
#### This allows for the combination of two dist plots for different variables for comparison

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
# x and y are the names of columns am plotting against each other
# data is the name of the data I want to plot from
# I can provide another argument kind= which can either be:
# scatter(default), hex, reg or kde. 
```

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```

### Categorical plots

#### Bar Plots
##### They are used to plot categorical data based on some calculations known as the estimator such as the mean, median etc against a numerical data. By default sns uses the mean to plot this. I can define a custom function of what I want and pass it to the estimator or even use functions from other libraries such as numpy

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# Here I want to use the standard deviation from numpy
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

#### Count Plots
##### This counts the number of occurence of the specified categorical column. It is therefore just like a barchat but just that the y axis is always by default the number of occurences of each category and as such, estimators cannot be used.

```python
sns.countplot(x='sex', data=tips)
```

#### Box and Violin Plots
##### They show the distribution of a categorical data against a numerical data

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex',
               dodge=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

### Matrix Plots - heat maps, cluster maps
#### The dataset for heat map must be in matrix form for it to work. I.e the column label must match the index label as shown in out[43] below

```python
# import second set of data from sns library
flights = sns.load_dataset('flights')
flights.head()
```

```python
tips.head()
```

```python
# To convert to a matrix we either change to a pivot table
# or get the correlation data
tipsMatrix = tips.corr()
```

```python
sns.heatmap(tipsMatrix, annot=True, cmap='coolwarm')
```

```python
flights
```

```python
# Using pivot tables
flightPivot = flights.pivot_table(index='month',columns='year',
                    values='passengers')
```

```python
sns.heatmap(flightPivot,cmap='magma',linecolor='white',
            linewidths=1)
```

```python
sns.clustermap(flightPivot,cmap='coolwarm',standard_scale=1)
```
