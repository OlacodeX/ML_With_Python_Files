```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)#11 linearly spaced numbers from 0-5
y = x ** 2
```

```python
x
```

```python
y
```

### First method: The functional method

```python
# functional method of using plt
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)#1 for no of rows, 2 for columns, 
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
# This allows for plotting 2 plots on a single canvas
# This makes sure there is spacing btwn the plots
# Always add it after your plots.
plt.tight_layout()
```

#### Second method:
##### object oriented method.

#### The Manual OOM

```python
# second method object oriented
# create a figure
fig = plt.figure()

# set axes to the figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])#left,bottom, height,width
# The values must be between 0 and 1
axes.plot(x,y) # Plot the axes
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
# Plot two graphs on canva
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('X label larger plot')
axes1.set_ylabel('Y label larger plot')
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_xlabel('X label subplot')
axes2.set_ylabel('Y label subplot')
axes2.set_title('Subplot Title')
```

##### The automatic OOM

```python
# Here I am using turple unpacking to generate sub plots
# Just like the functional method above
fig, axes = plt.subplots(nrows=1, ncols=2)
# the plt.subplots adds the axes automatically to the fig object
# depending on the nrows & ncolumns

# The returned axes is a list of iterable objects.
# This means I can use a for loop like this:
for axis in axes:
    axis.plot(x,y)
# This makes sure there is spacing btwn the plots
# Always add it after your plots.
plt.tight_layout()

#axes.plot(x,y)
```

```python
# I can also index it depending on the number of columns.
# As in the case above, I have two columns one row:
# So I can assess the axes by indexing like:
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

# This makes sure there is spacing btwn the plots
# Always add it after your plots.
plt.tight_layout()
```

### Figure Size, Aspect Ratio and DPI

```python
fig = plt.figure(figsize=(8,2), dpi=100)
# figsize accepts a turple where the first element is the width
# second is the height all in inches
# The dpi - dots per inch which you can leave at the default
# By not specifying it.

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# The above code is written like this for the shorter version
fig, axes = plt.subplots(figsize=(8,2))
# figsize accepts a turple where the first element is the width
# second is the height all in inches
# The dpi - dots per inch which you can leave at the default
# By not specifying it.

axes.plot(x,y)
```

```python
# Plotting two lines on a single graph
# Adding a legend to it
fig,axes = plt.subplots(figsize=(8,4))
axes.plot(x,x**2, label=' X Squared')
axes.plot(x,x**3, label='X Cube')
# The label argument is necessary for the legend method to work
# Add a legend
axes.legend(loc=0)
# loc can be specified to tell plt where to place the legend.
# it ranges from 0-10 where 0 is best location as plt determines
```

### Plot Appearance

```python
fig, axes = plt.subplots(figsize =(8,4))
axes.plot(x,y,color='red', linewidth=2,alpha=0.3,
         linestyle='-', marker='o', markersize=10,
         markerfacecolor='yellow',markeredgewidth=3,
         markeredgecolor='green')
# I can use lw as a shorthand for linewidth,
# alpha specifies the thickness of the line,
# I can use ls for linestyle.
# Markers marks each points in the dataset on the line.
# possible marker symbols: marker = '+', 'o', '*', 's', ',', 
# '.', '1', '2', '3', '4', ...
# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
```

### Setting the limit of values to be plotted

```python
fig, axes = plt.subplots(figsize =(8,4))
axes.plot(x,y,color='red', linewidth=2,alpha=0.3,
         linestyle='--', marker='o', markersize=10,
         markerfacecolor='yellow',markeredgewidth=3,
         markeredgecolor='green')
axes.set_xlim([0,1])
axes.set_ylim([0,2])
```
